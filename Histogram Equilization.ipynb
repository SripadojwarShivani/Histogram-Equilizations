```python
import cv2
```

```python
#reading an image
img=cv2.imread("/home/shivanisri/Downloads/light.jpeg")
cv2.imshow("before",img)
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
equalized = cv2.equalizeHist(gray)
```

```python
cv2.imshow("Result",equalized)
cv2.waitKey(0)
```
